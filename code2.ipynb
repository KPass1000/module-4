```python
#Custom Portfolio
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
aaple_data = pd.read_csv('aapl_historical.csv')
goog_data = pd.read_csv('goog_historical.csv')
cost_data = pd.read_csv('cost_historical.csv')
```

```python

combined_df = pd.concat([aaple_data, goog_data, cost_data], axis="columns", join="inner")
combined_df.sort_index(inplace=True)
combined_df.head()
```

```python
combined_df.drop(columns=['Symbol', 'Trade DATE'], inplace=True)
combined_df.head()
combined_df.columns = ['AAPL', 'GOOG', 'COST']
combined_df.head()
```

```python
daily_returns = combined_df.pct_change().dropna()
daily_returns.head()
```

```python
#Calculate Weighted Returns of Portfolio
my_returns = daily_returns
portfolio_weights = {'AAPL': 0.3, 'GOOG': 0.3, 'COST': 0.3}
weighted_returns = (my_returns * pd.Series(portfolio_weights)).sum(axis=1)
my_returns['Weighted Returns'] = weighted_returns
print (daily_returns)
```

```python
df = pd.DataFrame(daily_returns)
std_deviation = df.std(axis=0)
print ("Standard Deviation")
print (std_deviation)
```

```python
#Annualized Standard Deviation
import math
annual_std = 4.950053 * math.sqrt(252)
print ("AAPL")
print (annual_std)
annual_std = 11.728481 * math.sqrt(252)
print ("GOOG")
print (annual_std)
annual_std = 1.923349 * math.sqrt(252)
print ("COST")
print (annual_std)
```

```python
aapl_rolling_std = my_returns['AAPL'].rolling(window=21).std()
goog_rolling_std = my_returns['GOOG'].rolling(window=21).std()
cost_rolling_std = my_returns['COST'].rolling(window=21).std()
plt.figure(figsize=(10, 6))
plt.plot(aapl_rolling_std, label='AAPL')
plt.plot(goog_rolling_std, label='GOOG')
plt.plot(cost_rolling_std, label='COST')
plt.title('Rolling Standard Deviation (21-day Window)')
plt.xlabel('Date')
plt.ylabel('Standard Deviation')
plt.legend()
plt.show()
aapl_rolling_std.head()
```

```python
import seaborn as sns
correlation = daily_returns.corr()
correlation
sns.heatmap(correlation, vmin=-1, vmax=1)
```
